Watermarking adalah sebuah teknik untuk menyelipkan sebuah informasi ke dalam sebuah media tertentu, dalam hal ini adalah citra. Ada 3 buah macam watermarking yang biasa digunakan yaitu modifikasi LSB dan noise-based. Dalam tugas rancang ini saya menggunakan multibit watermark noise-based. Informasi yang akan saya selipkan adalah sebuah logo $32\times32$ (`logo.bmp`), sementara media citranya adalah sebuah gambar standar $512\times512$ (`lena512.bmp`). Berikut ini adalah diagram alir yang menjelaskan cara kerja multibit watermarking noise-based.

![caption](Diagram Alir 1.png)
<center> Gambar 1. Cara Kerja Multibit Watermarking Noise-based </center>

Hal pertama yang kita lakukan adalah mempartisi citra yang asli menjadi $1024$ kluster, dimana tiap kluster merupakan sebuah citra $16\times16$. Tiap bagian akan memuat sebuah informasi berupa bit logo yang bersangkutan. Jika bit logo adalah `1`, maka kluster yang bersangkutan akan kita beri _noise_, sebaliknya jika bit logo adalah `0`, maka kluster yang bersangkutan tidak kita beri _noise_ (atau dengan kata lain, kluster tersebut kita biarkan apa adanya). Tentu _noise_ yang kita gunakan merupakan hasil dari _pseudo-random_ generator, sehingga kita bisa atur pola _noise_ tertentu dengan menggunakan sebuah kunci tertentu. Artinya, kita akan mampu menghasilkan pola _noise_ yang sama jika kunci yang digunakan sama.

Selain itu, untuk mengekstrak logo dari citra ter-_watermark_, saya akan menggunakan _cross-correlation_ per kluster antara citra ter-_watermark_ dengan _noise_. Berikut ini adalah diagram alir yang menjelaskan cara kerja untuk mengekstrak logo dari citra ter-_watermark_.

<h2> Bab 2 Implementasi </h2>
</center>

---

Pertama, saya input terlebih dahulu citra yang akan digunakan

```matlab
ori = imread('lena512.bmp');
% imshow(ori);
```

```matlab
logo = imread('logo.bmp');
% imshow(logo);
```

Langkah berikutnya, saya atur gain _noise_ serta kunci yang akan digunakan

```matlab
amp = 100;
kunci = 1234;
```

Selanjutnya, saya akan membuat _noise_ $16\times16$ dengan menggunakan fungsi `randsrc()`

```matlab
noise = randsrc(16, 16, -amp:amp, kunci);
```

Kemudian, saya akan membuat citra kosong $512\times512$ yang nantinya menjadi tempat untuk menampung citra ter-_watermark_.

```matlab
ukuran = size(ori);
im_wm = uint8(zeros(ukuran));
```

Berikutnya, saya akan membagi citra asli menjadi beberapa kluster lalu kita isi dengan informasi yang bersangkutan beserta _noise_.

```matlab
for i = 1 : ukuran(1)
    for j = 1 : ukuran(2)
        x = mod(i, 16) + 1;
        y = mod(j, 16) + 1;
        if x == 1
            a = floor(i/16);
        else
            a = floor(i/16) + 1;
        end
        if y == 1
            b = floor(j/16);
        else
            b = floor(j/16) + 1;
        end        

        if logo(a,b) == 0
            im_wm(i, j) = ori(i, j) + (noise(x,y)*-1);
        else
            im_wm(i, j) = ori(i, j) + noise(x, y);
        end
    end
end
```

Mari kita lihat hasil citra ter-_watermark_ lalu membandingkannya dengan citra aslinya.

```matlab
subplot(121); imshow(ori);
title('ORI');
subplot(122); imshow(im_wm);
title('ORI + WM (100)');
```

Keterangan:
* ORI = Citra asli.
* ORI + WM (100) = Citra ter-_watermark_ dengan gain 100.

Perlu diakui, saya memasang gain terlalu besar sehingga citra menjadi sangat terdistorsi berat. Tentu kita harus mencari gain yang tepat agar kualitas citra terjaga, salah satunya dengan menganalisis _peak signal-to-noise ratio_ atau sering disingkat menjadi PSNR. Definisi umum PSNR adalah sebagai berikut

$$ \text{PSNR} = 10\log\left( \frac{\text{peakval}^2}{\text{MSE}} \right) $$

Citra yang digunakan merupakan citra 8 bit, sehingga $\text{peakval} = 2^8 = 255$. Kemudian $MSE$ adalah singkatan dari _mean square error_ antara citra asli dengan citra ter-_watermark_. Dalam MATLAB sudah tersedia fungsi khusus untuk menghitung PSNR, yaitu `peaksnr = psnr(A, ref)`. Variabel `A` adalah citra yang akan hendak kita cari PSNR-nya, sementara variable `ref` adalah citra acuan yang kita pakai. Sebagai contoh, kita akan mencoba menghitung psnr citra ter-_watermark_ sebelumnya.

```matlab
peaksnr = psnr(im_wm, ori)
```

Umumnya, semakin tinggi PSNR akan semakin bagus pula kualitas citra (artinya semakin mirip citra ter-_watermark_ dengan citra asli). Pada bab berikutnya, saya akan berbicara lebar mengenai PSNR tersebut.

Selanjutnya, kita akan mencoba mengekstrak logo dari citra terwatermark tersebut. Langkah pertama adalah mengatur kunci yang digunakan untuk menghasilkan pola _noise_ yang serupa dengan proses _watermarking_ sebelumnya.

```matlab
amp = 100;
kunci = 1234;
noise = randsrc(16, 16, -amp:amp, kunci);
```

Kemudian, kita akan membuat citra kosong $32\times32$ yang nantinya menjadi tempat untuk menampung logo hasil ekstrak dari citra ter-_watermark_.

```matlab
res = zeros(32,32);
```

Seperti sebelumnya, citra ter-_watermark_ akan saya bagi menjadi beberapa buah kluster, lalu tiap kluster akan dicocokkan dengan _noise_ (menggunakan operasi _cross-correlation_). Idenya adalah jika semakin tinggi nilai _cross-correlation_ maka kita bisa berasumsi bahwa bit logo pada kluster tersebut adalah `1`, jika sebaliknya maka bit logo pada kluster tersebut adalah `0`. Namun tentu kita butuh sebuah nilai _threshold_ untuk menentukan apakah bit logo yang disimpan adalah `0` atau `1`.

```matlab
res = zeros(32, 32);
for i = 0 : 31
    for j = 0 : 31
        kx = i*16;
        ky = j*16;
        sample = im_wm(kx+1 : kx+16, ky+1 : ky+16);
        r = normxcorr2(noise, sample);
        res(i+1, j+1) = r(16,16);
    end
end
```

Sekarang logo `res` sudah berisi nilai _cross-correlation_. Sekarang kita amati pada baris pertama logo `res` dengan logo asli-nya untuk menentukan nilai _threshold_ yang tepat.

```matlab
plot(logo(1,:)); hold on;
plot(res(1,:)); hold off;
```

Ternyata hasilnya cukup baik. Disini kita bisa lihat bahwa untuk kasus bit logo `1`, nilai _cross-correlation_ konsisten di bawah 0, sebaliknya untuk kasus bit logo `0`, nilai _cross-correlation_ konsisten di atas 0. Kita bisa anggap bahwa nilai _threshold_ yang tepat adalah 0. Kita jalankan lagi perintah pada cell (..) namun sudah menggunakan nilai _threshold_ untuk meng-_generate_ logo.

```matlab
res = zeros(32, 32);
for i = 0 : 31
    for j = 0 : 31
        kx = i*16;
        ky = j*16;
        sample = im_wm(kx+1 : kx+16, ky+1 : ky+16);
        r = normxcorr2(noise, sample);
        res(i+1, j+1) = r(16,16);
        if res(i+1, j+1) < 0 
            res(i+1, j+1) = 1;
        else
            res(i+1, j+1) = 0;
        end
    end
end
res = logical(res);
```

Sekarang, kita siap untuk membandingkan antara logo asli dengan logo hasil ekstrak.

```matlab
subplot(121); imshow(logo);
title('LOGO');
subplot(122); imshow(res);
title('LOGO EXT');
```

Keterangan:
* LOGO = logo yang asli
* LOGO EXT = logo hasil ekstrak dari citra ter-_watermark_

Logo yang diekstrak ternyata sangat menyerupai dengan logo aslinya. Tentu dengan catatan ada beberapa bit yang _miss_ di beberapa tempat, namun secara keseluruhan sudah cukup bagus.

<h2> Bab 3 Pemilihan Gain yang Tepat </h2>
</center>

---

Sebelumnya, kita sudah melihat bahwa teknik multibit watermarking noise-based sudah cukup baik untuk menyimpan informasi. Namun, kita berandai-andai apakah adalah pemilihan gain yang tepat supaya kita bisa mengurangi distorasi pada citra ter-_watermark_?

Pertama, saya akan menghitung terlebih dahulu PSNR citra ter-_watermark_. Seperti yang sudah saya sampaikan di bab terdahulu bahwa PSNR menentukan kualitas citra. Bisa kita katakan PSNR adalah parameter 'kemiripan' antara citra ter-_watermark_ dengan citra asli. Semakin mirip citra ter-_watermark_ dengan citra asli, maka nilai PSNR akan semakin tinggi.

Kedua, saya akan menghitung sebuah parameter yang saya beri nama 'likeness' (atau dengan kata lain kemiripan). Parameter ini mengukur kemiripan antara logo hasil ekstrak dengan logo asli. Berikut ini adalah definisi yang saya buat.

$\text{likeness} = \frac{\Sigma\;\text{bit yang cocok}}{\Sigma\;\text{keseluruhan bit}}$

Jika kita perhatikan sebenarnya parameter `likeness` ini adalah kebalikan dari _error bit_. Definisinya

$\text{error bit} = \frac{\Sigma\;\text{bit yang tidak cocok}}{\Sigma\;\text{keseluruhan bit}}$

Sekarang mari kita coba untuk ambil sampel 100 buah titik sampel denga nilai gain yang bervariasi dari 0.5 sampai 50.

```matlab
clear; clc;

ori = imread('lena512.bmp');
% imshow(ori);

logo = imread('logo.bmp');
% imshow(logo);

test = 0.5 : 0.5 : 50;
likeness = zeros(1, length(test));
psnr_val = zeros(1, length(test));
```

```matlab
for c = 1 : length(test)
        
    amp = test(1, c);
    kunci = 1234;
    
    noise = randsrc(16, 16, -amp:amp, kunci);
    
    ukuran = size(ori);
    im_wm = uint8(zeros(ukuran));
    
    for i = 1 : ukuran(1)
        for j = 1 : ukuran(2)
            x = mod(i, 16) + 1;
            y = mod(j, 16) + 1;
            if x == 1
                a = floor(i/16);
            else
                a = floor(i/16) + 1;
            end
            if y == 1
                b = floor(j/16);
            else
                b = floor(j/16) + 1;
            end        

            if logo(a,b) == 0
                im_wm(i, j) = ori(i, j) + (noise(x,y)*-1);
            else
                im_wm(i, j) = ori(i, j) + noise(x, y);
            end
        end
    end    

    res = zeros(32, 32);
    for i = 0 : 31
        for j = 0 : 31
            kx = i*16;
            ky = j*16;
            sample = im_wm(kx+1 : kx+16, ky+1 : ky+16);
            r = normxcorr2(noise, sample);
            res(i+1, j+1) = r(16,16);
            if res(i+1, j+1) < 0 
                res(i+1, j+1) = 1;
            else
                res(i+1, j+1) = 0;
            end
        end
    end
    res = logical(res);

    likeness(1, c) = (sum(sum(not(xor(logo, res))))*100)/(32*32);
    psnr_val(1, c) = psnr(im_wm, ori);

end
```

```matlab
plot(likeness);
xlabel('sampel ke-n');
ylabel('Likeness Logo (%)');
yyaxis right;
plot(psnr_val);
ylabel('PSNR Citra Ter-watermark (dB)');
```

Dari grafik di atas kita ketahui bahwa,
* Semakin tinggi gain _noise_, maka semakin rendah nilai PSNR. Tentu saja itu berarti citra ter-_watermark_ akan semakin terdistorsi.
* Semakin tinggi gain _noise_, maka semakin tinggi likeness logo. Tentu saja itu berarti error bit yang mungkin pada logo hasil ekstrak akan semakin kecil.

Kita perhatikan pula bahwa PSNR paling tinggi berada di atas `50 dB`, sementara nilai likeness logo paling tinggi bisa mencapai `95%` lebih. Dengan grafik demikian, kita ketahui bahwa kita tidak bisa mencari PSNR yang tinggi sembari menjaga nilai likeness logo juga tinggi, karena keduanya memiliki tren yang berlawanan. Oleh karenanya, yang bisa kita lakukan adalah mencari _trade-off_ yang paling optimum, yaitu pada persinggung antara fungsi `likeness logo` dengan `PSNR`. Namun sebelumnya, kita harus menormalisasi terlebih dahulu grafik tersebut.

```matlab
plot(likeness/likeness(1,end));
xlabel('sampel ke-n');
ylabel('Likeness Logo (%)');
axis([1 100 0.4 1]);
yyaxis right;
plot(psnr_val/psnr_val(1,1));
ylabel('PSNR Citra Ter-watermark (dB)');
axis([1 100 0.4 1]);
```

Kita peroleh titik paling optimum ada pada sampel ke-10, yang itu berarti memiliki nilai gain _noise_ sebesar 5.
